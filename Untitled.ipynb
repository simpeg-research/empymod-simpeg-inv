```python
import numpy as np
import matplotlib.pyplot as plt

import discretize
from SimPEG import maps
import empymod
```

```python
depth = np.array([ 0, 1000, 1010, 1020]) # Layer boundaries
res = np.array([1e8, .3, 1, 0.01, 1])    # Air, sea, overburden, target, underburden
src_z = 980
src = [0, 0, src_z, 0, 90]                   # Src coordinates
rec = [1, 0, src_z, 0, 90]  # Rec coordinates
times = np.logspace(-5, 1, 20)

data = empymod.loop(src, rec, depth, res, times, mrec=True, signal=-1)
```

```python
fig, ax = plt.subplots(1, 2)

def plot_model(depth, res, ax):
    depth_plot = np.vstack([depth, depth]).flatten(order="F")[1:]
    depth_plot = np.hstack([depth_plot, depth_plot[-1]*1.5])
    res_plot = np.vstack([res, res]).flatten(order="F")[2:]
    return ax.semilogx(res_plot, depth_plot, label="model")
    
plot_model(depth, res, ax[0])
ax[0].set_ylim([1100, 900])
ax[1].loglog(times, data, "-o", label="data")
 

for a in ax:
    a.legend()
    a.grid()

plt.tight_layout()
```

```python
from SimPEG.simulation import BaseSimulation 
from SimPEG.survey import BaseSrc, BaseRx, BaseSurvey
from SimPEG import props
```

```python
class EmpymodSimulation(BaseSimulation):
    """
    Problem class for a simulation conducted with empymod
    """
    
    rho, rho_map, rho_deriv = props.Invertible(
        "Electrical Resistivity (Ohm m)"
    ) 
    
    deleteTheseOnModelUpdate = ['_Japprox']
    
    def __init__(self, mesh, survey, rho_map=None, rho=None):
        assert mesh.dim == 1, "only 1D modelling supported"
        super().__init__(mesh=mesh)
        self.survey = survey
        self.rho_map = rho_map
        self.rho = rho
        
        # set air at infinity
        depth = self.mesh.gridN.copy()
        depth[0] = -np.inf
        self.depth = depth
    
    def dpred(self, m):
        """
        Computes the fields
        """
        
        # set the model (this performs the mappings)
        self.model = m
        data = empymod.loop(self.src, self.rec, self.mesh, self.rho, self.times, mrec=True, signal=-1)
        return data
    
    def Japprox(self, m, perturbation=0.1, min_perturbation=1e-3):
        """
        Approximate sensitivity computed using a finite difference approach
        """
        if getattr(self, '_Japprox', None) is None:
            self.model = m 
            delta_m = min_perturbation # np.max([perturbation*m.mean(), min_perturbation])

            J = []

            for i, entry in enumerate(m):
                mpos = m.copy()
                mpos[i] = entry + delta_m
                
                mneg = m.copy()
                mneg[i] = entry - delta_m
                
                pos = self.dpred(self.rho_map * mpos)
                neg = self.dpred(self.rho_map * mneg)
                J.append((pos - neg) / (2.*delta_m))

            self._Japprox = np.vstack(J).T
            
        return self._Japprox
    
    def Jvec(self, m, v, f=None):
        """
        Sensitivity times a vector
        """
        self.model = m
        return self.Japprox(m).dot(v)

    def Jtvec(self, m, v, f=None):
        """
        Adjoint Sensitivity times a vector
        """
        self.model = m
        return self.Japprox(m).T.dot(v)
```

```python
mesh = discretize.TensorMesh([np.diff(depth)]) 
simulation = EmpymodSimulation(mesh, times, src, rec, rho_map=maps.ExpMap(mesh))
```
